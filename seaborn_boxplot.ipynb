First we have to import few libraries before we start plotting the data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

Now we have to read the dataset . In our case the dataset we are using is Titanic dataset. To read the dataset we will use sns.load_dataset.

```python
titanic = sns.load_dataset('titanic')
```

To display data we will use head function which will print the first five rows for us 

```python
titanic.head()
```

Now we will plot box plot using seaborn.Boxplot for one variable

```python
sns.boxplot( titanic['fare'])
```

Boxplot for two variables

```python
sns.boxplot( y = titanic['age']  , x = titanic['sex'])
```

Plot the distribution using hue:

```python
sns.boxplot( y = titanic['age']  , x = titanic['sex'] , hue = titanic['class'])
```

Plot the distribution on horizontal axis:

```python
sns.boxplot( x = titanic['pclass']  ,y = titanic['embarked']  , orient = 'horizontal')
```

Plot swarmplot

```python
sns.boxplot( y = titanic['age']  , x = titanic['embarked'] , palette = 'winter_r')
sns.swarmplot( y = titanic['age']  , x = titanic['embarked'] , color = '0.8' )
```
