First we have to import few libraries before we start plotting the data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

Now we have to read the dataset . In our case the dataset we are using is Titanic dataset. To read the dataset we will use sns.load_dataset.

```python
titanic = sns.load_dataset('titanic')
```

To display data we will use head function which will print the first five rows for us 

```python
titanic.head()
```

Now we will plot Violinplot using seaborn.

```python
sns.violinplot(x = 'age' , data = titanic)
```

violinplot for two variables

```python
sns.violinplot(x = 'age'  , y = 'class' , data = titanic)
```

Now we will be using hue parameter to add some more data and make it more interesting.

```python
sns.violinplot(x = 'age'  , y = 'class' , data = titanic , hue = 'sex')
```

Now to change the color we will use palette parameter

```python
sns.violinplot(x = 'embarked'  , y = 'survived' , data = titanic , hue = 'class' , palette = "RdBu")
```

Using split parameter

```python
sns.violinplot(x = 'age'  , y = 'class' , data = titanic , hue = 'sex' , split = True)
```

To add horizontal line

```python
sns.violinplot(x = 'class'  , y = 'age' , data = titanic , hue = 'sex' , split = True , inner = 'quartile')
```

Scaling the plot

```python
sns.violinplot(x = 'class'  , y = 'age' , data = titanic , hue = 'sex', 
               inner = 'quartile' , scale = 'count')
```

plotting each data point as horizontal line

```python
sns.violinplot(x = 'class'  , y = 'age' , data = titanic , hue = 'sex', 
               inner = 'stick' , scale = 'count')
```

to scale the hue

```python
sns.violinplot(x = 'class'  , y = 'age' , data = titanic , hue = 'sex', 
               inner = 'stick' , scale = 'count' , scale_hue = False)
```

```python
sns.violinplot(x = 'class'  , y = 'age' , data = titanic , hue = 'sex', 
               inner = 'stick' , scale = 'count' , scale_hue = False , split = True)
```

To change the bandwidth

```python
sns.violinplot(x = 'class'  , y = 'age' , data = titanic , hue = 'sex', 
               inner = 'stick' , scale = 'count' , scale_hue = False , split = True , bw = 0.1)
```
