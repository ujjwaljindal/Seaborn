First we have to import few libraries before we start plotting the data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

Now we have to read the dataset . In our case the dataset we are using is Titanic dataset. To read the dataset we will use sns.load_dataset.

```python
titanic = sns.load_dataset('titanic')
```

To display data we will use head function which will print the first five rows for us 

```python
titanic.head()
```

Now we will plot strip plot using seaborn.Boxplot for one variable

```python
sns.stripplot( x = 'age' , data = titanic)
```

stripplot for two variables

```python
sns.stripplot( x = 'age' ,y = 'class' , data = titanic )
```

Plot the distribution using hue:

```python
sns.stripplot( x = 'age' ,y = 'class' , hue = 'sex' ,  data = titanic , palette = 'husl' )
```

Plot the distribution on vertical axis:

```python
sns.stripplot( y = 'age' ,x = 'class' , hue = 'sex' ,  data = titanic , palette = 'husl' )
```

To make it less dispersed use the jitter parameter

```python
sns.stripplot( y = 'age' ,x = 'class' , hue = 'sex' ,  data = titanic , palette = 'husl' , jitter=False )
```

we can also pass a number to jitter parameter. Higher the number more dispersed it is.

```python
sns.stripplot( y = 'age' ,x = 'class' , hue = 'sex' ,  data = titanic , palette = 'husl' , jitter = 0.2 )
```

We can also add outlines to data points

```python
sns.stripplot( y = 'age' ,x = 'class' , hue = 'sex' ,  data = titanic , palette = 'husl' , linewidth = .5 )
```

To split the hue we will use dodge parameter

```python
sns.stripplot( y = 'age' ,x = 'class' , hue = 'sex' ,  data = titanic , palette = 'husl' , dodge = True )
```

To change the shape of datapoints we will use marker parameter

```python
sns.stripplot( y = 'age' ,x = 'class' , hue = 'sex' ,  data = titanic , palette = 'husl' , dodge = True  , marker = 'D')
```

To change the size of datapoints we will use size parameter

```python
sns.stripplot( y = 'age' ,x = 'class' , hue = 'sex' ,  data = titanic , palette = 'husl' , dodge = True  , marker = 'D' , size = 8)
```

To change the color of edges 

```python
sns.stripplot( y = 'age' ,x = 'class' , hue = 'sex' ,  data = titanic , palette = 'husl' , dodge = True  , marker = 'D' ,
              size = 10, edgecolor = 'yellow' , alpha = 0.5)
```

plotting stripplot and boxplot together

```python
sns.stripplot( y = 'age' ,x = 'class' ,  data = titanic , palette = 'husl' )
sns.boxplot( y = 'age' ,x = 'class'  ,  data = titanic , palette = 'husl' )
```

Plotting stripplot and violinplot together

```python
sns.stripplot( y = 'age' ,x = 'class' ,  data = titanic , palette = 'husl' )
sns.violinplot( y = 'age' ,x = 'class'  ,  data = titanic , color = '0.9' )
```
