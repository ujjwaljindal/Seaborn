First we have to import few libraries before we start plotting the data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

Now we have to read the dataset . In our case the dataset we are using is iris dataset. To read the dataset we will use sns.load_dataset.

```python
iris = sns.load_dataset('iris')
```

To display data we will use head function which will print the first five rows for us 

```python
iris.head()
```

Now we will plot PAIRGRID using seaborn.

```python
x= sns.PairGrid( iris)
x.map(plt.scatter)
```

Now we will be changing diagonal in histogram.

```python
x= sns.PairGrid( iris)
x = x.map_diag(plt.hist)
```

Now to add other grids we will use offdaig function

```python
x= sns.PairGrid( iris)
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

Now we will use hue parameter

```python
x= sns.PairGrid( iris , hue = 'species')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

now we will add legend

```python
x= sns.PairGrid( iris , hue = 'species')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

to change the color we will use palette

```python
x= sns.PairGrid( iris , hue = 'species' , palette = 'coolwarm')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

To change the histogram type

```python
x= sns.PairGrid( iris , hue = 'species')
x = x.map_diag(plt.hist , histtype = 'step')
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

To see only few features we will use vars parameter

```python
x= sns.PairGrid( iris , vars = ['petal_length' , 'petal_width'])
x = x.map(plt.scatter)
```

using many parameters

```python
x= sns.PairGrid( iris , hue = 'species' ,  vars = ['petal_length' , 'petal_width'])
x = x.map_diag(plt.hist , histtype = 'step')
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

To use different x and y 

```python
x= sns.PairGrid( iris , hue = 'species' ,  x_vars = ['petal_length' , 'petal_width'] , y_vars = ['sepal_length' , 'sepal_width'])
x = x.map(plt.scatter)
x = x.add_legend()
```

To display different plots above and below the diagonals

```python
x= sns.PairGrid( iris , hue = 'species')
x = x.map_diag(plt.hist , histtype = 'step')
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)

x = x.add_legend()
```

To use different markers

```python
x= sns.PairGrid( iris , hue = 'species' ,palette = 'coolwarm',  hue_kws = {'marker':['D' , 's' , '+']})
x = x.map(plt.scatter)
x = x.add_legend()
```
